```python
import pandas as pd
import numpy as np
import scipy as sp
```

```python
data=pd.read_csv('data/fj1.csv', dtype={"resolution": str, "gsubname2":str, "gname3":str, "gsubname3":str, "claimmode3_txt":str, "weaptype4_txt":str, "weapsubtype4_txt":str, "divert":str, "kidhijcountry":str})
```

```python
data = data[data.doubtterr == 0]
data.info()
```

```python
new_feature = ['nkill', 'nkillter', 'nwound', 'nwoundte']
pd_feature = ['region','crit1', 'crit2', 'crit3', 'success', 'suicide', 'attacktype1', 'targtype1', 'nperps', 'weaptype1', 'ransom']

data['nkill'].fillna(0, inplace = True)
data['nkillter'].fillna(0, inplace = True)
data['nwound'].fillna(0, inplace = True)
data['nwoundte'].fillna(0, inplace = True)

data['kill_and_wound'] = (data['nkill'] - data['nkillter'])*0.7 + (data['nwound'] - data['nwoundte'])*0.3
```

```python
data.loc[ data['kill_and_wound'] <= 0, 'kill_and_wound'] = 1
data.loc[ (data['kill_and_wound'] > 0) & (data['kill_and_wound'] <= 10), 'kill_and_wound'] = 2
data.loc[ (data['kill_and_wound'] > 10) & (data['kill_and_wound'] <= 50), 'kill_and_wound'] = 3
data.loc[ (data['kill_and_wound'] > 50) & (data['kill_and_wound'] <= 200), 'kill_and_wound'] = 4
data.loc[ data['kill_and_wound'] > 200, 'kill_and_wound'] = 5
```

```python
data_pd = pd.DataFrame({
        'kill_and_wound': data['kill_and_wound'],
    })

for feature in pd_feature:
    data_pd[feature] = data[feature]
```

```python
data_pd['nperps'].fillna(5, inplace = True)
data_pd.loc[ (data_pd['nperps'] >= 1) & (data_pd['nperps'] < 10), 'nperps'] = 1
data_pd.loc[ (data_pd['nperps'] >= 10) & (data_pd['nperps'] < 100), 'nperps'] = 2
data_pd.loc[ (data_pd['nperps'] >= 100) & (data_pd['nperps'] < 500), 'nperps'] = 3
data_pd.loc[ data_pd['nperps'] >= 500, 'nperps'] = 4
data_pd.loc[ data_pd['nperps'] == -99, 'nperps'] = 5
data_pd['nperps'] = data_pd['nperps'].astype(int)

data_pd['nperps'].fillna(5, inplace = True)
data_pd['nperps'] = data_pd['nperps'].astype(int)
data_pd.loc[ (data_pd['nperps'] >= 1) & (data_pd['nperps'] < 10), 'nperps'] = 1
data_pd.loc[ (data_pd['nperps'] >= 10) & (data_pd['nperps'] < 100), 'nperps'] = 2
data_pd.loc[ (data_pd['nperps'] >= 100) & (data_pd['nperps'] < 500), 'nperps'] = 3
data_pd.loc[ data_pd['nperps'] >= 500, 'nperps'] = 4
data_pd.loc[ data_pd['nperps'] == -99, 'nperps'] = 5

data_pd.loc[ data_pd['ransom'] == -9, 'ransom'] = 3
data_pd['ransom'].fillna(3, inplace = True)
```

```python
data_pd.to_csv('data/data_pd2.csv', index=False)
```
